```python
chemicals =['boron', 'calcium', 'clay', 'copper', 'ec_salts',
       'exchangeable_acidity', 'iron', 'magnesium', 'manganese', 'phosphorus',
       'potassium', 'sand', 'silt', 'sodium', 'sulphur', 'zinc', 'ph']
```

```python
def validate_intersections(chemical):
    curr_train = pd.read_csv(os.path.join(PATH_TO_CURR_SPLITS, f"{chemical}_train_sample_codes.csv"))
    curr_test = pd.read_csv(os.path.join(PATH_TO_CURR_SPLITS, f"{chemical}_test_sample_codes.csv"))
    curr_valid = pd.read_csv(os.path.join(PATH_TO_CURR_SPLITS, f"{chemical}_valid_sample_codes.csv"))
    
    prev_train = pd.read_csv(os.path.join(PATH_TO_PREV_SPLITS, f"{chemical}_train_sample_codes.csv"))
    prev_test = pd.read_csv(os.path.join(PATH_TO_PREV_SPLITS, f"{chemical}_test_sample_codes.csv"))
    prev_valid = pd.read_csv(os.path.join(PATH_TO_PREV_SPLITS, f"{chemical}_valid_sample_codes.csv"))
    
    
    current = pd.DataFrame()
    
    current = pd.concat([curr_train, curr_test, curr_valid])
    
    for sample in current['x'].values:
        if str(sample).strip() in prev_train['x'].str.strip().values:
            print(sample)
            n = pd.DataFrame()
            n['x'] = [sample]
            curr_train = pd.concat([curr_train, n])
#             curr_train.to_csv(os.path.join(PATH_TO_CURR_SPLITS, f"{chemical}_train_sample_codes.csv"))
        if str(sample).strip() in prev_test['x'].str.strip().values:
            print(sample)
            n = pd.DataFrame()
            n['x'] = [sample]
            curr_train = pd.concat([curr_train, n])
#             curr_test.to_csv(os.path.join(PATH_TO_CURR_SPLITS, f"{chemical}_test_sample_codes.csv"))
        if str(sample).strip() in prev_valid['x'].str.strip().values:
            print(sample)
            n = pd.DataFrame()
            n['x'] = [sample]
            curr_train = pd.concat([curr_train, n])
#             curr_test.to_csv(os.path.join(PATH_TO_CURR_SPLITS, f"{chemical}_valid_sample_codes.csv"))
    
  
```

```python
for chemical in chemicals:
    validate_intersections(chemical)
```
